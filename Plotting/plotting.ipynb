```python
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np

df_dqn = pd.read_csv('dqn_ep_rew_mean.csv')
df_a2c = pd.read_csv('a2c_ep_rew_mean.csv')
df_ddpg = pd.read_csv('ddpg_ep_rew_mean.csv')
```

```python
df_dqn_cut = df_dqn.drop(df_dqn[df_dqn['Step'] > 46000].index)
df_dqn_cut
```

```python
df_a2c_cut = df_a2c.drop(df_a2c[df_a2c['Step'] > 46000].index)
df_a2c_cut
```

```python
df_ddpg_cut = df_ddpg.drop(df_ddpg[df_ddpg['Step'] > 46000].index)
df_ddpg_cut
```

```python
plt.plot(df_dqn_cut['Step'], df_dqn_cut['Value'], label='DQN')
plt.plot(df_a2c_cut['Step'], df_a2c_cut['Value'], label='A2C')
plt.plot(df_ddpg_cut['Step'], df_ddpg_cut['Value'], label='DDPG')
plt.plot(df_dqn_cut['Step'], [40] * len(df_dqn_cut['Step']), label='Threshold')
plt.xlabel('Total Time Step')
plt.ylabel('Mean Reward of Each Episode')
plt.legend(loc='upper left', ncol=2)

plt.show()
```
